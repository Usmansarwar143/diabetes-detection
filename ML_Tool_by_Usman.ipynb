```python
import pandas as pd
import joblib
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
#Reading the data from CSV File
data=pd.read_csv('C:/Users/Dell/Downloads/diabetes.csv')
```

```python
# See first  rows of data to understand the format and ensure the accuance of data
data.head()
```

```python
#DATA CLEANING
#To check if there any null value exist in the data
data.isna().sum().sum()
```

```python
# TO list all the columns of data, including feature and target
data.columns
```

```python
#features in the data
X=data[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']]
#target
Y=data['Outcome']
```

```python
 #Splitting the data for training and testing the Model
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
X_test.head()
```

```python
#Usage of a preprocessing techniqe for better output
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
# Calling and Training of the  Model (Logistic Regression)
model=svm.SVC(kernel='linear')
model.fit(X_train,y_train)
```

```python
#Let the trained Model predict The test value of X
y_pred=model.predict(X_test)
print(y_pred)
y_test=np.array(y_test)
y_test=y_test[np.newaxis, :]
print(y_test)
```

```python
#Checking the Accuracy os the Model
a_s=accuracy_score(y_test,y_pred)
```

```python
print("The Accuracy Score is: ",a_s)
```
